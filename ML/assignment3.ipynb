```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv("/content/salary_data.csv")
```

```python
df
```

```python
# prepare
x = df["YearsExperience"].values.reshape(-1,1)
y = df["Salary"].values.reshape(-1,1)
```

```python
# standardscaler without library func
x = (x - np.mean(x)) / np.std(x)
x
```

```python
# init
m = len(y)
theta_0 = 0  # Bias
theta_1 = 0  # Weight
alpha = 0.01  # Learning rate
epochs = 1000
```

```python
# linear regression with gradient descent
l = []
for epoch in range(epochs):
  y_pred = theta_0 + theta_1 * x
  loss = (1/(2*m)) * np.sum((y_pred - y) ** 2)
  l.append(loss)

  # gradient descent
  d_theta_0 = (1/m) * np.sum(y_pred -y)
  d_theta_1 = (1/m) * np.sum((y_pred -y) * x)

  # update param
  theta_0 -= alpha * d_theta_0
  theta_1 -= alpha * d_theta_1

  # plot hypothesis func
  if epoch % 100 == 0:
    plt.scatter(x,y,color='red',label='Data Points')
    plt.plot(x,y_pred,color='blue',label=f'Epoch {epoch}')
    plt.xlabel('years of experience')
    plt.ylabel('salary')
    plt.legend()
    plt.show()
```

```python
# final prediction
y_pred_final = theta_0 + theta_1 * x
```

```python
r2 = r2_score(y,y_pred_final)
print(f'r2 score: {(r2):.2f}')
```

```python
# plot
plt.plot(range(epochs), l, color='blue')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss Curve')
plt.show()
```
